### 예측 변수와 시계열 예측

- 설명 모형(explanatory model) 

지역의 시간당 전기 수요(ED, electricity demand)를 예측할때, 전기 수요의 변동에 영향을 주는 예측 변수(predictor variable)을 사용해서 예측

\begin{align*}
  \text{ED} = & f(\text{현재 기온, 경제 상황, 인구,}\\
 &  \qquad\text{시간, 요일, 오차}).
\end{align*}


- 시계열 모델

전기 수요 데이터가 시계열이라서, 이전 시점의 데이터만을 사용할 수 있음.
아래는 시계열 예측식 

\begin{align*}
   \text{ED}_{t+1} = f(\text{ED}_{t}, \text{ED}_{t-1}, \text{ED}_{t-2}, \text{ED}_{t-3},\dots, \text{오차}),
\end{align*}
  
t 는 현재 시간,  t + 1 은 다음 시간,  t − 1 은 이전 시간, t − 2 는 2시간 전, ...

- 혼합된 모델(mixed models)

위의 두 모델의 특징을 결합한 세 번째 종류의 모델

\begin{align*}
  \text{ED}_{t+1} = f(\text{ED}_{t}, \text{현재 기온, 시간, 요일, 오차}).
\end{align*}

동적 회귀(dynamic regression) 모델, 패널(panel) 데이터 모델, 종단적(longitudinal) 모델, 수송 함수(transfer function) 모델, ( 
f 가 선형이라고 가정하는) 선형 시스템(linear system) 모델으로 불려짐.

## 1.6 예측 작업의 기본 단계

예측 작업에는 보통 다섯 가지 기본 단계

- 1단계: 문제 정의.

신중하게 문제를 정의하려면, 예측값이 어떤 방식으로 사용될 지를 이해하는 것, 누가 예측값을 요청하는지, 예측 함수가 예측을 필요로 하는 회사의 요구와 얼마나 잘 맞는지가 필요

- 2단계: 정보 수집.

항상 적어도 두 종류의 데이터가 필요: (1) 통계 데이터, 그리고 (2) 데이터를 모으고 예측값을 사용하는 사람의 누적된 전문 지식

- 3단계: 예비(탐험적; exploratory) 분석.

항상 데이터를 그래프로 나타내는 것으로 시작


- 4단계: 모델을 선택하고 맞추기.

과거 데이터를 이용할 수 있는 정도, 예측 변수(forecast variable)와 설명 변수(explanatory variable) 사이 관계의 강도, 그리고 예측값이 사용될 방식에 따라 사용할 가장 좋은 모델이 달라짐


- 5단계: 예측 모델을 사용하고 평가하기.

### 시계열의 빈도

“빈도(frequency)”는 계절성 패턴이 반복되기 전까지의 관측값의 수


| 데이터  | 빈도 |
| :----: | :---: |
| 연     | 1   |
| 분기   | 4    |
| 월     | 12    |
| 주     | 52    |

관측값의 빈도가 1주일에 한 번보다 클 때는, 빈도(frequency)를 다루는 방법

- 일별 관측값으로 구성된 데이터에는 주별 계절성(빈도=7)이나 연간 계절성(빈도=365.25)
- 1분마다 관측된 데이터에는 시간별 계절성(빈도=60), 일별 계절성(빈도=24x60=1440), 주별 계절성(빈도=24x60x7=10080), 연간 계절성(24x60x365.25=525960) 등이 있음.

### 2.3 시계열 패턴

***추세(trend)***

    데이터가 장기적으로 증가하거나 감소할 때, 추세(trend)가 존재
    추세를 “변화 방향”이라고 함.

***계절성(seasonality)***

    해마다 어떤 특정한 때나 1주일마다 특정 요일에 나타나는 것 같은 계절성 요인이 시계열에 영향을 줄 때 계절성(seasonality) 패턴
    계절성은 항상 알려진 상수 빈도 형태

***주기성(cycle)***

    고정된 빈도가 아닌 형태로 증가나 감소하는 모습을 보일 때 주기(cycle)
    보통 이러한 요동은 경제 상황 때문에 일어나고, 흔히 “경기 순환(business cycle)”
    요동의 지속기간은 적어도 2년 이상

## 2.4 계절성 그래프

계절성 그래프(seasonal plot)는 각 “계절(season)”에 대해 관측한 데이터를 나타낸다는 점만 제외하고는 시간 그래프와 비슷

```r
ggseasonplot(a10, year.labels = TRUE, year.labels.left = TRUE) +
  ylab("백만 달러") +
  xlab("월") +
  ggtitle("계절성 그래프: 당뇨병 약 판매량")
```

극좌표(polar coordinate)를 사용하여 계절성 그래프를 바꿔 나타내면 도움이 됩니다. 아래에서 볼 수 있는 것처럼, polar=TRUE로 두면, 수평축 대신 원형축으로 나타냅니다.

```r
ggseasonplot(a10, polar = TRUE) +
  ylab("백만 달러") +
  xlab("월") +
  ggtitle("계절성 극좌표 그래프: 당뇨병 약 판매량")
```

## 2.8 자기상관

상관값이 두 변수 사이의 선형 관계의 크기를 측정하는 것처럼, 자기상관(autocorrelation)은 시계열의 시차 값(lagged values) 사이의 선형 관계를 측정.

시차 그래프(lag plot)에서 각 패널과 관련된 몇 가지 자기상관 계수가 있습니다.  예를 들면, $$ r_{1} 은 y_{t} 와 y_{t-1} 사이의 관계를 측정하고,   r_{2} 은 y_{t} 와 y_{t-2} 관계 $$ 

\begin{align*}
r_{k} = \frac{\sum\limits_{t=k+1}^T (y_{t}-\bar{y})(y_{t-k}-\bar{y})}
 {\sum\limits_{t=1}^T (y_{t}-\bar{y})^2},
\end{align*}

여기에서  T  는 시계열의 길이입니다.


참고 : 상관계수(correlation coefficient)
\begin{align*}
    r = \frac{\sum (x_{t} - \bar{x})(y_{t}-\bar{y})}{\sqrt{\sum(x_{t}-\bar{x})^2}\sqrt{\sum(y_{t}-\bar{y})^2}}.
\end{align*}



    다음의 표는 맥주 생산량 데이터에 대한 처음 9개의 자기상관(autocorrelation) 계수를 보여줍니다.
    자기상관 계수는 보통 자기상관함수(ACF)를 나타내기 위해 그립니다. 이 그래프는 상관도표(correlogram)라고도 알려져 있습니다.

```r
ggAcf(beer2)
```

이 그래프에서:

   - r4 는 다른 시차에 대한 경우보다 높습니다. 데이터에서 나타나는 계절성 패턴 때문에 그렇습니다: 고점은 4개의 분기가 떨어져 있는 경향이 있고 저점은 4개의 분기가 떨어져 있는 경향이 있습니다.
   - r2 는 저점이 고점에 대해 2개의 분기가 떨어져 있는 경향이 있어서 다른 시차에 대한 경우보다 더 큰 음수입니다.
   - 파란 점선은 상관값이 0보다 유의미하게 큰 지를 나타냅니다. 이러한 것을 2.9 절에서 설명합니다.

#### ACF 그래프에서 추세와 계절성

- 데이터에 추세가 있을 때, 시간적으로 가까운 관측값이 크기도 비슷하기 때문에 작은 뒤처짐에 대한 자기상관은 크면서 양의 값을 갖는 경향
- 추세가 있는 시계열의 ACF에는 뒤처짐들이 증가할 때 느리게 감소하는 양의 값을 갖는 경향
- 데이터에 계절성이 있을 때, (계절성 빈도의 배수로 나타나는) 계절성 뒤처짐에 대한 자기상관이 다른 시차에 대한 경우보다 클 것
    
데이터에 추세와 계절성이 둘 다 있을 때는, 이러한 효과의 조합을 볼 수 있음.

```r
aelec <- window(elec, start=1980)
autoplot(aelec) + xlab("연도") + ylab("기가와트시(GWh)")
```

```r
ggAcf(aelec, lag=48)
```

## 2.9 백색잡음

자기상관(autocorrelation)이 없는 시계열을 백색잡음(white noise)이라고 부름.

```r
set.seed(30)
y <- ts(rnorm(50))
autoplot(y) + ggtitle("백색잡음")
```

```r
ggAcf(y)
```

- 백색잡음 시계열에 대해, 자기상관 값 각각이 거의 0일 것이라고 기대
- 백색잡음 시계열에 대해, ACF에서 뾰족한 막대의 95%가 $$ \pm 2/ \sqrt{T}$$ 에 들어갈 것을 기대
- T = 50  이라서 $$ \pm 2/\sqrt{50} = \pm 0.28 $$ 에 경계
