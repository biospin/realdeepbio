## 현대 생명과학은 데이터에 관한 것이다

- 생명과학은 근본적인 본질이 바뀌었으며, 로봇 공학 및 소형화 된 실험을 통해 **실험 데이터 생성량이 극적으로 증가**함.

- 거대한 데이터 세트를 생성 할 수있는 유전자 시퀀싱과 같은 실험은 값 싸고 일상적으로 진행됨.

- **유전자 시퀀싱**의 진보로 인해 당뇨병, 암 및 낭포 성 섬유증과 같은 **유전 질환**을 비롯한 수많은 건강 관련 결과에 대해 개인의 유전자 코드를 연결하는 **데이터베이스**가 구축되었음.

- 이 데이터를 분석하고 마이닝으로 계산하는 기술을 사용하여 과학자들은 이러한 **질병의 원인**을 이해하고 이러한 이해를 사용하여 **새로운 치료법**을 개발한다.

- 인간 관측에 주로 의존했던 분야는 이제 수동으로 분석 할 수 없는 데이터 세트를 활용하고 있다.

- 기계 학습은 이제 **세포 이미지를 분류**하는 데 일상적으로 사용된다

- 이러한 기계 학습 모델의 결과는 **암 종양**을 **확인**하고 **분류**하고 잠재적인 **질병 치료의 효과**를 평가하는 데 사용된다.

- 실험 기술의 진보는 **화학 물질의 구조**와 이러한 화학 물질이 광범위한 **생물학적 과정 또는 활동에 미치는 영향**을 카탈로그 화하는 여러 **데이터베이스의 개발**로 이어진다.

- 이러한  **structure–activity relationships (SAR)** 는 cheminformatics 또는 해당 분야의 기초를 형성한다. 과학자들은 이러한 대형 데이터 세트를 연구하고 **차세대 약물 개발**을 주도 할 **예측 모델**을 구축하기 위해 이 데이터를 사용한다.

- 많은 양의 데이터를 통해 과학과 전산 영역 모두에서 편한 **새로운 종류의 과학자**가 필요하다. 이러한 **하이브리드 기능**을 가진 사람들은 대규모 데이터 세트의 구조와 경향을 열어 내일의 과학적 발견을 가능하게 할 수 있다.

## 본 책에서 다룰 내용

첫 부분에 딥러닝에 대한 개요와 그것이 생명 과학에 어떻게 적용될 수 있는지에 대해 설명. 컴퓨터 학습은 "데이터로 학습 할 수 있도록 컴퓨터 프로그래밍의 과학 (및 예술)"으로 정의됨.

2 장에서는 딥러닝에 대한 간략한 소개를 제공. 기계 학습이 선형 회귀와 같은 간단한 작업을 수행하고 생명 과학의 실제 문제를 해결하는 데 일반적으로 사용되는 보다 정교한 모델로 진행하는 데 어떻게 사용될 수 있는지에 대한 사례부터 시작한다. 일반적으로 기계 학습은 데이터 집합을 모델을 생성하는 데 사용되는 훈련 집합과 모델의 성능을 평가하는 데 사용되는 테스트 집합으로 분할하여 진행된다. 2 장에서는 예측 모델의 훈련과 유효성 확인을 둘러싼 몇 가지 세부 사항을 논의한다. 모델이 생성되면 하이퍼 매개 변수로 알려진 여러 특성을 변경하여 성능을 일반적으로 최적화 할 수 있다. 이 장에서는이 프로세스의 개요를 제공한다. 딥러닝은 하나의 기법이 아니라 관련된 일련의 기법이다. 2 장에서는 가장 중요한 딥러닝 변수에 대해 소개한다.

3 장에서는 다양한 생명 과학 응용 프로그램을 위한 딥러닝 모델 생성을 단순화하도록 특별히 설계된 오픈 소스 프로그래밍 라이브러리 인 DeepChem을 소개하고 개요를 제공 한 후, 분자의 독성을 예측하기위한 모델을 생성하기 위해 DeepChem 라이브러리를 사용하는 방법을 보여주는 프로그래밍 예제를 소개한다. 다음 DeepChem이 현대 생물학에서 공통적인 작업인 이미지를 분류하는 데 어떻게 사용될 수 있는지 보여준다. 암 진단부터 녹내장 검출에 이르기까지 다양한 영상 응용 분야에서 딥러닝이 사용된다.

4 장에서는 기계 학습을 분자에 적용하는 방법에 대한 개요를 제공한다. 우리는 분자들, 우리 주위의 모든 것을 구성하는 요소들을 소개함으로써 시작한다. 분자는 빌딩 블록과 유사하지만 유연하고 동적인 행동을 나타낸다. 딥러닝과 같은 계산 방법을 사용하여 분자를 특성화하기 위해서는 컴퓨터에서 분자를 나타내는 방법을 찾아야 한다. 이러한 인코딩은 이미지를 픽셀 집합으로 표현할 수 있는 방식과 유사하다. 후반부에서는 분자를 표현할 수 있는 많은 방법과 어떻게 딥러닝 모델을 구축하는 데 사용될 수 있는지 설명합니다.

5 장에서는 물리 법칙을 생물 현상에 적용하는 생물 물리학 분야에 대해 소개한다. 생명을 가능하게하는 분자 머신인 단백질에 대한 논의부터 시작한다. 약물이 신체에 미치는 영향을 예측하는 핵심 구성 요소는 단백질과의 상호 작용을 이해하는 것이다. 이러한 효과를 이해하기 위해 단백질 구성 방법과 단백질 구조가 어떻게 다른지에 대한 개요로 시작한다. 단백질은 3D 구조가 생물학적 기능을 지시하는 개체이다. 기계 학습 모델이 약물 기능이 단백질 기능에 미치는 영향을 예측하기 위해서는 기계 학습 프로그램에서 처리 할 수있는 형태로 3D 구조를 표현해야합니다. 후반부에 단백질 구조가 표현 될 수있는 많은 방법을 모색한다. 이 지식을 바탕으로 약물 분자가 단백질과 상호 작용하는 정도를 예측하기 위해 딥러닝을 사용하는 또 다른 코드 예제를 검토한다.

유전학은 현대 의학의 핵심 구성 요소가 되었다. 종양의 유전자 시퀀싱은 암의 맞춤 치료를 가능하게하고 의학 혁명의 잠재력을 가지고 있다. 예전에는 거대한 투자를 필요로하는 복잡한 과정이었던 유전자 시퀀싱이 일상화되어 일상적으로 수행 될 수 있다. 우리는 강아지 주인들이 저렴한 애완 동물의 혈통을 결정하는 유전자 검사를 받을 수있는 시점까지 도달했다. 6 장에서는 단백질을 생산하는 데 사용되는 DNA와 RNA에 대한 소개부터 시작하여 유전학과 유전체학에 대한 개요를 제공한다. 최근의 발견에 따르면 DNA와 RNA의 상호 작용은 처음에는 믿어지는 것보다 훨씬 더 복잡하다. 후반부에서는 DNA와 RNA의 상호 작용에 영향을 미치는 여러 요인을 예측하는 데 얼마나 깊은 학습이 사용될 수 있는지 보여주는 몇 가지 코드 예제를 제시한다.

앞부분에서 우리는 생물학 및 의료 영상 분석에 대한 심화 학습의 적용을 통해 얻은 많은 발전에 대해 암시했다. 이 실험에서 연구 된 많은 현상은 너무 작아서 인간의 눈으로 관찰 할 수 없다. 딥러닝 방법으로 사용되는 이미지를 얻으려면 현미경을 사용해야 한다.
7 장은 학교에서 사용 된 간단한 광학 현미경부터 원자 분해능으로 이미지를 얻을 수있는 정교한 도구에 이르기까지 다양한 형태의 현미경에 대한 개요를 제공한다. 이 장에서는 또한 현재 접근법의 몇 가지 한계를 다루고 딥러닝 모델을 유도하는 이미지를 얻는 데 사용되는 실험 파이프 라인에 대한 정보를 제공한다.

유망한 영역은 의학 진단에 대한 딥러닝의 적용이다. 의학은 믿을 수 없을 정도로 복잡하며, 의사는 개인이 이용할 수있는 의학 지식을 모두 구현할 수 없다. 이상적인 상황에서 기계 학습 모델은 의학 문헌을 소화하고 의료 전문가가 진단을 내리는 데 도움을 줄 수 있다. 아직 이 지점에 도달하지는 못했지만 많은 긍정적인 단계가 이루어지고 있다.

8장은 의료 진단을 위한 기계 학습 방법의 역사에서 시작하여 수작업으로 부호화 된 규칙에서 의료 결과의 통계 분석으로의 전환을 차트로 보여준다. 앞서 논의한 많은 주제와 마찬가지로 핵심 구성 요소는 기계 학습 프로그램에서 처리 할 수있는 형식으로 의료 정보를 나타낸다. 이 장에서는 전자 건강 기록 및 이 기록을 둘러싼 몇 가지 문제점에 대해 소개한다. 대부분의 경우 의료 이미지는 매우 복잡하며 이러한 이미지의 분석 및 해석은 숙련 된 전문가에게조차 어려울 수 있다. 이러한 경우 딥러닝은 이미지를 분류하고 주요 기능을 식별함으로써 인간 분석가의 기술을 향상시킬 수 있다. 8장은 다양한 영역에서 의료 영상을 분석하는 데 얼마나 깊은 학습이 사용되는지에 대한 많은 예를 통해 결론을 맺고있다.

이전에 언급했듯이 기계 학습은 약물 발견 노력의 핵심 구성 요소가 되고 있다. 과학자들은 약물 분자와 단백질 간의 상호 작용을 평가하기 위해 딥러닝 모델을 사용합니다. 이러한 상호 작용은 환자에게 치료 효과가 있는 생물학적 반응을 유도 할 수 있다. 지금까지 논의한 모델은 차별 모델이다. 분자의 특성 집합이 주어지면,이 모델은 어떤 특성의 예측을 생성한다. 이러한 예측은 입력 분자를 필요로 하는데, 이는 사용 가능한 분자의 대규모 데이터베이스로부터 파생되거나 과학자의 상상력에서 비롯 될 수 있다. 현재 존재하는 것 또는 우리가 상상할 수있는 것에 의지하기보다는 새로운 분자를 "발명"할 수있는 컴퓨터 프로그램이 있다면 어떨까? 9 장은 생성 모델 (generative model)이라고 불리는 일종의 딥러닝 프로그램을 제시한다. GAN은 처음에는 기존 분자 집합에 대해 훈련된 다음 새로운 분자를 생성하는 데 사용된다. 이러한 분자를 생성하는 딥러닝 프로그램은 또한 새로운 분자의 활동을 예측하는 다른 모델에 의해 영향을 받을 수 있다.

지금까지 우리는 딥러닝 모델을 "블랙 박스"로 논의했다. 모델에 일련의 입력 데이터를 제공하고 모델은 예측이 생성 된 방법이나 이유에 대한 설명없이 예측을 생성한다. 이러한 유형의 예측은 많은 경우 최적 이하일 수 있다. 우리가 의학적 진단을 위한 깊이있는 학습 모델을 가지고 있다면 진단의 근거가되는 추론을 이해할 필요가 있다. 진단의 이유에 대한 설명은 의사에게 예측에 대한 더 많은 확신을 주며 치료 결정에도 영향을 줄 수 있다. 딥러닝에 대한 하나의 역사적인 단점은 모델이 종종 신뢰할 만하지만 해석하기 어려울 수 있다는 사실이었다. 예측을 유도한 요인을 사용자가 더 잘 이해할 수 있도록 다양한 기술이 현재 개발 중이다. 10장에서는 모델 예측에 대한 인간의 이해를 돕기 위해 사용되는 몇 가지 기법에 대해 간략히 설명한다. 예측 모델의 또 다른 중요한 측면은 모델의 예측 정확성이다. 모델 정확도를 이해하면 해당 모델에 얼마나 의존해야하는지 판단 할 수 있다. 잠재적으로 생명을 구하는 진단을 위해 기계 학습을 사용할 수 있다고 가정 할 때 모델 정확도를 이해하는 것이 중요하다. 10장의 마지막 절에서는 모델 예측의 정확성을 평가하는 데 사용할 수있는 몇 가지 기술에 대해 간략히 설명한다.

11 장에서는 DeepChem을 사용한 실제 사례 연구를 제시한다. 이 예에서는 가상 스크리닝이라는 기술을 사용하여 신약 발견의 잠재적 출발점을 식별한다. 약물 발견은 종종 선별 검사 (screening)로 알려진 기술로 시작하는 복잡한 과정이다.
스크리닝은 결국 약물을 생성하기 위해 최적화 될 수있는 분자를 확인하는데 사용된다. 스크리닝은 실험적으로 수행 할 수 있다. 여기에는 수백만 개의 분자가 분석으로 알려진 소형 생물학적 테스트 또는 가상 스크린을 사용하는 컴퓨터에서 테스트된다. 가상 스크리닝에서 알려진 약물 또는 기타 생물학적 활성 분자 세트가 기계 학습 모델을 학습하는 데 사용된다. 이 기계 학습 모델은 큰 분자 세트의 활동을 예측하는 데 사용된다. 기계 학습 방법의 속도로 인해 컴퓨터 시간의 며칠 만에 수억 개의 분자를 처리 할 수 있다.

이 책의 마지막 장에서는 생명 과학의 깊은 학습의 현재 영향과 미래 잠재력에 대해 살펴본다. 데이터 세트의 가용성 및 품질을 비롯한 현재의 노력에 대한 여러 가지 과제가 논의된다. 우리는 또한 진단, 맞춤 의학, 제약 개발 및 생물학 연구를 포함한 여러 다른 영역에서 기회와 잠재적 위험을 강조한다.
